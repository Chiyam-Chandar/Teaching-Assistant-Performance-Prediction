```python
# change the filename/path if you downloaded the zip and unzipped
import pandas as pd

# example if you downloaded CSV; replace with actual path
df = pd.read_csv('teaching_assistant_data.csv')   # or the actual filename
df.head(8)
```

```python
# replace 'Class_attribute' by the exact column name for performance
# The original code caused a KeyError because 'Class_attribute', 'Native_teacher', and 'Semester' were not actual column names.
# Based on the df.columns and df.describe() output, I've mapped the likely intended columns.
print(df['3.1'].value_counts())
print(df['3.1'].value_counts(normalize=True).round(3))
# Example for other categorical columns:
print(df['1'].value_counts())
print(df['1.1'].value_counts())
```

```python
missing = df.isnull().sum().sort_values(ascending=False)
missing[missing>0]
```

```python
# Unique counts to spot typos
for col in df.columns:
    print(col, "->", df[col].nunique(), "unique values")
    if df[col].nunique() < 30:     # only print value counts for small cardinality columns
        print(df[col].value_counts())
        print()
```

```python
summary_text = []
summary_text.append(f"Shape: {df.shape}")
summary_text.append("Columns: " + ", ".join(df.columns))
summary_text.append("Missing values:\n" + str(missing[missing>0].to_dict()))
summary_text.append("Target class counts:\n" + str(df['3.1'].value_counts().to_dict()))

with open('data_summary.txt','w') as f:
    f.write("\\n\\n".join(summary_text))

print("Saved data_summary.txt")
```
